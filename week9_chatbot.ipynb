```python
from openai import OpenAI
import os

client = OpenAI(
    base_url="https://openrouter.ai/api/v1",
    api_key=os.getenv("OPENROUTER_API_KEY")
)

response = client.chat.completions.create(
    model="openai/gpt-4o-mini",  # OpenRouter model format
    messages=[
        {"role": "user", "content": "Say hello and introduce yourself"}
    ],
    temperature=0.7,
    max_tokens=200
)

print(response.choices[0].message.content)
```

# PART 2: BUILD BASIC CHATBOT

## Task 2.1: Create Conversation Loop

```python
def chat(messages):
    """
    Send conversation to model and return response
    """
    try:
        response = client.chat.completions.create(
            model="openai/gpt-3.5-turbo",  # OpenRouter format  # openai/gpt-4o-mini
            messages=messages,
            max_tokens=500,
            temperature=0.7
        )
        return response.choices[0].message.content

    except Exception as e:
        return f"❌ Error: {str(e)}"


# =========================
# Conversation Setup
# =========================
messages = [
    {"role": "system", "content": "You are a helpful assistant."}
]

print("🤖 Chatbot ready! Type 'quit' to exit.\n")

# =========================
# Chat Loop
# =========================
while True:
    user_input = input("You: ")

    if user_input.lower().strip() == "quit":
        print("👋 Goodbye!")
        break

    # Add user message
    messages.append({"role": "user", "content": user_input})

    # Get response
    assistant_response = chat(messages)

    # Add assistant message
    messages.append({"role": "assistant", "content": assistant_response})

    print(f"Assistant: {assistant_response}\n")
```

## Task 3.1: Create HR Assistant

```python
# HR system prompt
hr_system_prompt = (
    "You are an HR assistant for a technology company. "
    "You help employees understand company policies in a clear and professional way.\n\n"

    "Company policies:\n"
    "- Vacation: 15 days per year\n"
    "- Sick leave: Unlimited (with manager approval)\n"
    "- Remote work: 3 days per week\n"
    "- Health insurance: Fully covered\n"
    "- 401(k) matching: Up to 5%\n\n"

    "Your responsibilities:\n"
    "1. Answer employee questions about policies\n"
    "2. Be friendly, supportive, and professional\n"
    "3. If unsure, suggest contacting the HR department\n"
    "4. Keep responses concise (2–3 sentences)"
)
```

```python
messages = [
    {
        "role": "system",
        "content": hr_system_prompt
    }
]
```

```python
def chat(messages):
    response = client.chat.completions.create(
        model="openai/gpt-3.5-turbo",
        messages=messages,
        temperature=0.5,
        max_tokens=300
    )
    return response.choices[0].message.content

# =========================
# Chat Loop
# =========================
while True:
    user_input = input("You: ")

    if user_input.lower().strip() == "quit":
        print("👋 Goodbye!")
        break

    # Add user message
    messages.append({"role": "user", "content": user_input})

    # Get response
    assistant_response = chat(messages)

    # Add assistant message
    messages.append({"role": "assistant", "content": assistant_response})

    print(f"Assistant: {assistant_response}\n")
```
